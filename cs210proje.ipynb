```python
file_name = "Cleaned_Netflix_and_Spotify_Merged_Data.csv"
merged_data_cleaned = pd.read_csv(file_name)

mood_comparison = merged_data_cleaned[["Mood_x", "Mood_y"]]
mood_comparison["Mood_x"] = mood_comparison["Mood_x"].str.split(", ")
mood_comparison["Mood_y"] = mood_comparison["Mood_y"].str.split(", ")

mood_combinations = mood_comparison.explode("Mood_x").explode("Mood_y")

mood_counts = mood_combinations.value_counts().reset_index()
mood_counts.columns = ["Netflix Mood", "Spotify Mood", "Frequency"]

pivot_table = mood_counts.pivot(index="Netflix Mood", columns="Spotify Mood", values="Frequency").fillna(0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="coolwarm", cbar=True)
plt.title("Effect of Combined Netflix Moods on Spotify Moods")
plt.xlabel("Spotify Mood")
plt.ylabel("Netflix Mood")
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder

file_name = "Cleaned_Netflix_and_Spotify_Merged_Data.csv"
merged_data_cleaned = pd.read_csv(file_name)

X = merged_data_cleaned["Mood_x"]
y = merged_data_cleaned["Mood_y"]

le_x = LabelEncoder()
le_y = LabelEncoder()
X_encoded = le_x.fit_transform(X)
y_encoded = le_y.fit_transform(y)

X_encoded = X_encoded.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

model = BalancedRandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Classification Report:")

unique_classes = sorted(set(y_train).union(set(y_test)))

print(classification_report(
    y_test,
    y_pred,
    labels=unique_classes,
    target_names=le_y.inverse_transform(unique_classes),
    zero_division=0
))

print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")

def recommend_mood(netflix_mood):
    netflix_mood_encoded = le_x.transform([netflix_mood])

    predicted_mood = model.predict([netflix_mood_encoded])

    return le_y.inverse_transform(predicted_mood)[0]

recommendation = recommend_mood("Joyful/Relaxed")
print(f"Recommended Spotify Mood: {recommendation}")
```

	•	Joyful/Relaxed and Mysterious/Tense classes had no representation in the test set, leading to scores of 0.00 for all metrics.
